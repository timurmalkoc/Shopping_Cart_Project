```python
from IPython.display import clear_output
import time
d = {"Apple":[5, 1.5]}

# menu section -------------------------------------------------------------------------------------
def print_menu():
    # Printing Menu
    option = input("Add an item in the list press .......... 1 \n"+
                   "Update an item from the list press ..... 2 \n"+
                   "Remove an item from the list press ..... 3 \n"+
                   "Print the list press ................... 4 \n"+
                   "Claer the list press ................... 5 \n"+
                   "Finish shopping press .................. 6 \n")
    return option

# info menu -----------------------------------------------------------------------------------------
def info():
    print("""
    ******************************************
    *                                        *
    *      Shopping List APP version 1.0     *
    *                                        *
    *           By Timur Celik               *
    *                                        *
    *           ☆┌─┐  ─┐   ☆                *
    *            │▒│ /▒/                     *
    *            │▒│/▒/                      *
    *            │▒ /▒/─┬─┐   ◯             *
    *            │▒│▒|▒│▒│                   *
    *           ┌┴─┴─┐-┘─┘                   *
    *           │▒┌──┘▒▒▒│◯                 *
    *           └┐▒▒▒▒▒▒┌┘                   *
    *           ◯└┐▒▒▒▒┌                    *
    *                                        *
    ******************************************
    """)

# Adding Item Section --------------------------------------------------------------------------------
def add_item(item=""):
    if item=="":
        item = input("Enter item name .... ")
    item = item.strip().title() #Capitalizing first letter and removing spaces
    
# item exist in the list
    if item in d:
        print("Item is in the list !")
        update_item(item)
         
# Adding new item in the list
    else:
        # getting quantity
        q=0
        p=0
        while True:
            clear_output(wait=True)
            q = input(f"Enter "+item+"'s quantity ... ")
            if q.isnumeric() and int(q)>0:
                clear_output(wait=True)
                break
            else:
                print("PLEASE ENTER ITEM's Quantity GREATER THAN 0 OR A NUMBER")
                
        # getting price
        while True:
            clear_output(wait=True)
            p = input("Enter "+item +"'s price ... ")
            if p.replace('.','').isnumeric() and float(p)>0:
                clear_output(wait=True)
                d[item]=[int(q),float(p)]
                print(f"{item} is added with quantity {d[item][0]} and price ${d[item][1]:0.2f}") 
                break
            else:
                print("PLEASE ENTER ITEM GREATER THAN 0 OR A NUMBER")



# update Item Section ----------------------------------------------------------------------------------
def update_item(item=""):
    if item=="":
        item = input("Enter item name .... ")
    item = item.strip().title() #Capitalizing first letter    
    
# checking item is in the list --------------------------- update()
    if item in d:
        while True:
            clear_output(wait=True)
            c = input("Update only quantity press ..... 1\n"+
                      "Update only price .............. 2\n"+
                      "Update both price and quatity .. 3\n"+
                      "Return to main menu ............ 4\n")
            if c in {'1','2','3','4'}:
                break
            else:
                print("PLEASE ENTER A VALID SELECTION!")

# updating quantity index 0 ------------------------------- update()
        if c == '1':
            
            while True:
                clear_output(wait=True)
                q = input(f"Old Quantitiy is {d[item][0]} \n"+
                          f"Enter {item}'s new quantity ... ")
                if q.isnumeric() and int(q)>0:
                    d[item][0]=int(q)
                    clear_output(wait=True)
                    print(f"{item}'s quatity is updated {d[item][0]}")
                    break
                else:
                    print("PLEASE ENTER ITEM's QUANTITY GREATER THAN 0 OR A NUMBER")
                    
                    
# updating item price index 1 ------------------------------- update()
        if c == '2':
            
            while True:
                clear_output(wait=True)
                q = input(f"Old price is {d[item][1]} \n"+
                          f"Enter {item}'s new price ... ")
                if q.replace('.','').isnumeric() and float(q)>0:
                    d[item][1]=float(q)
                    clear_output(wait=True)
                    print(f"{item}'s price is updated ${d[item][1]:0.2f}")
                    break
                else:
                    print("PLEASE ENTER ITEM's PRICE GREATER THAN 0 OR A NUMBER")
                    
                    
# updating item quantitiy and price index 1 and 2 ------------ update()
        if c == '3':
            while True:
                clear_output(wait=True)
                q = input(f"Old Quantitiy is {d[item][0]} \n"+
                          f"Enter {item}'s new quantity ... ")
                if q.isnumeric() and int(q)>0:
                    d[item][0]=int(q)
                    clear_output(wait=True)
                    print(f"{item}'s quatity is updated {d[item][0]}")
                    break
                else:
                    print("PLEASE ENTER ITEM's QUANTITY GREATER THAN 0 OR A NUMBER")
                    
            while True:
                clear_output(wait=True)
                q = input(f"Old price is {d[item][1]} \n"+
                          f"Enter {item}'s new price ... ")
                if q.replace('.','').isnumeric() and float(q)>0:
                    d[item][1]=float(q)
                    clear_output(wait=True)
                    print(f"{item} is updated new quantity {d[item][0]} and new price ${d[item][1]:0.2f}") 
                    break
                else:
                    print("PLEASE ENTER ITEM's PRICE GREATER THAN 0 OR A NUMBER")
                       
                    
# item is not in the list so redirecting to adding new item funtion with item parameter
    else:
        add_item(item)
        
# Remove item from a list ---------------------------------------------------------------------------------
def remove_item():
    if item=="":
        item = input("Enter item name .... ")
    item = item.title() #Capitalizing first letter
    
    if item in d:
        del d[item]
        print(item+" is successfully removed from the list")
    
    else:
        print("THERE IS NO SUCH ITEM IN THE LIST !!!")

        
# Listing shoping the list --------------------------------------------------------------------------------
def print_list():
    if not len(d):
        clear_output(wait=True)
        print("List is empty !!!")
    else:
        clear_output(wait=True)
        print("-------------- ReceipT ---------------------\n")
        for item in d:
            print(f"{item}\t quantity {d[item][0]} \tunit price ${d[item][1]:0.2f} \titem(s) price ${d[item][0]*d[item][1]:0.2f}\n")
        print("********************************************")
        print(f"Number of item(s) {get_total()[1]} and Total amount ${get_total()[0]:0.2f}")

        
# Sub funciton - calculates total price in the list
def get_total():
    total = [0,0]
    for item in d:
        total[0]+=(d[item][0]*d[item][1])
        total[1]+=1
    return total


# Main funtion --------------------------------------------------------------------------------------------
def main():
    opt = '0' #keeps menu option
    clear_output(wait=True)
    info()
    time.sleep(5)
    clear_output(wait=True)
    while opt!='6':
        clear_output(wait=True)
        opt = print_menu()
        if not opt in {'1','2','3','4','5','6'}:
            print("PLEASE ENTER A VALID SELECTION!")
        else:
            if opt == '1':
                add_item()
            elif opt == '2':
                update_item()
            elif opt == '3':
                remove()
            elif opt == '4':
                print_list()
            elif opt == '5':
                d.clear()
                print("List is clered !!!")
main()    
```
